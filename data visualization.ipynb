```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.arange(20,50)
y=np.arange(30,60)
```

```python
plt.plot(x);
```

"LINE PLOT"(both one and two) 
Here we had plotted only one column value so it is used to find the trend(increasing or decreasing), seasonality

```python
plt.plot(x, y);
```

Here we had plotted 2 column features to find the relationship between both. The above graph shoes as x increases y also increases

```python
plt.bar(x, y, color = ['g', 'b'], align='center');
plt.xlabel('X')
plt.title('bar graph')
```

"BAR GRAPH"(two columns)

```python
plt.hist(x, bins=5);
```

"HISTOGRAM"(one column)
Here only one clounm feature can be inputed, It is used to find the density of a value between a range of value. For example:- if number 5 is repeated more time, using histogram we can identify

```python
plt.boxplot(x, patch_artist='red');
```

"BOX PLOT"(one column)
Here only one clounm feature can be inputed, It is used to find the outliers.
PERCENTILE is applied in Box Plot. 

```python
plt.subplot(2,3,1)
plt.plot(x);
plt.subplot(2,3,2)
plt.plot(y);
plt.subplot(2,3,3)
plt.boxplot(x);
plt.subplot(2,3,4)
plt.boxplot(y);
plt.subplot(2,3,5)
plt.plot(x,y);
plt.subplot(2,3,6)
sizes = [300,1099]
plt.pie(sizes, explode=(0.2,0), colors=['red','green'], shadow=True);
```

```python
sizes = [440, 1335]
plt.pie(sizes, explode=(0.2,0), colors=['red','green'], shadow=True, autopct='%f%%');
```

# seaborn

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.corr() #applied to find co-relation between datas
```

```python
sns.heatmap(df.corr(), fmt="f", cmap="Blues", annot=True);
plt.title('co-relation')
```

"HEATMAP"(all numeric columns)
It is used to find relation between datas in a DataFrame, here numeric values can only applied

.corr() and .heatmap is very good combination used for feature selection and EDA

```python
sns.jointplot(x='size', y='total_bill', data=df, kind='hex');
```

"JOINT PLOT"(two columns)
Probablity density function is used here, 2 column features are applied in join plot...
want to know more to understand this

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg');
```

```python
sns.pairplot(df);
```

"PAIR PLOT"(whole df is used)
Permutation and Combination of all values in a DataFrame

```python
sns.distplot(df['total_bill'], hist=True, kde=True, bins=10);
```

"DIST PLOT"(one column)
It is used to find how features are distributed in a column

### Categorical plot
boxplot, violinplot, countplot, barplot

```python
sns.countplot('sex', data=df, hue='smoker');
```

"COUNT PLOT"
It is used to find the density between each category in a column feature. For example:- the above graph shows that male are more than female

```python
sns.barplot(x='size', y='total_bill', data=df, hue='smoker');  #x is category_variable and y is numeric_variable
```

```python
sns.boxplot(x='day', y='tip', data=df, palette='rainbow', hue='sex');
```

"BOX PLOT"(two column)
percentile is applied
hue is used to find the relationship between categories

```python
sns.violinplot(x='day', y='total_bill', data=df, hue='smoker')
```
